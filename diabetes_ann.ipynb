```python
import numpy as np
import pandas as pd
import tensorflow as tf
df=pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
print(df.isna().sum())
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.fit_transform(xtest)
```

```python
from keras.src.layers.serialization import activation
ann=tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(8,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(8,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(xtrain,ytrain,batch_size=40,epochs=100)
```

```python
xtrain.shape
```

```python
pred=ann.predict(scaler.transform([[6,148,72,35,0,33.6,0.627,50]]))
if pred>0.5:
  print(1)
else:
  print(0)
```

```python
ann.summary()
```

```python
ann.evaluate(xtest,ytest)
```

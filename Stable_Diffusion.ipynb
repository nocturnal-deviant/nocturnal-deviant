```python
from diffusers import StableDiffusionPipeline,EulerDiscreteScheduler
#stable diffuson pipeline ---
#euler discret scheduler ---
import torch
```

```python
model_id='stabilityai/stable-diffusion-2'
```

```python
scheduler=EulerDiscreteScheduler.from_pretrained(model_id,subfolder='scheduler')

pipe=StableDiffusionPipeline.from_pretrained(model_id,scheduler=scheduler,revision='fp16',torch_dtype=torch.float16)
```

```python
pipe=pipe.to('cuda')
```

```python
prompt='Artificial Intelligence in healthcare'
```

```python
def text_to_imge(prompt):
  image=pipe(prompt,guidance_scale=10,height=768,width=768).images[0]
  return image
```

```python
text_to_imge(prompt)
```

```python
import gradio as gr
gr.Interface(text_to_imge,gr.Text(),gr.Image(),title='Stable Diffusion').launch()
```

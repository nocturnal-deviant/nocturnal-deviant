```python
import numpy as np
import pandas as pd
import tensorflow as tf
df=pd.read_csv('/content/lung_cancer_.csv')
df.head()
```

```python
df=df.drop(['Name','Surname'],axis=1)
```

```python
print(df.isna().sum())
```

```python
df.describe(include='all')
```

```python
# help(tf)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x.head()
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.fit_transform(xtest)
```

```python
from keras.src.layers.serialization import activation
#Initialization ANN
ann=tf.keras.models.Sequential()
```

```python
#adding hidden layers
ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
#2nd hidden layer
ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
#out put layer
ann.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
#compiling neural network
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(xtrain,ytrain,batch_size=10,epochs=100)
```

```python
xtrain.shape
```

```python
pred=ann.predict(scaler.transform([[30,0,5,2]]))
if(pred>0.5):
  print(1)
else:
  print(0)
```
